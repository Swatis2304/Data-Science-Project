```python
#keep the original ZORI data
#alter ZHVINTA2 to get the mean and sum of each year
ZORI2 = ZORI.copy()
```

```python
# Define the range of years you want to sum values for
start_year = 2015
end_year = 2023

# Iterate over each year
for year in range(start_year, end_year + 1):
    # Create a boolean mask for columns that start with the specific year
    year_cols = [col for col in ZORI2.columns if col.startswith(str(year))]

    # Sum the values of these columns for each row and create a new column for the sum
    ZORI2[f'Sum_{year}'] = ZORI2[year_cols].sum(axis=1)

    # Count non-NaN values for each row across the year-specific columns
    count_non_nan = ZORI2[year_cols].count(axis=1)

    # Calculate the mean and create a new column for it
    ZORI2[f'Mean_{year}'] = ZORI2[f'Sum_{year}'].div(count_non_nan)
```

```python
ZORI2['zipcode'] = ZORI2['RegionName']
```

```python
# List of specific columns you want to keep
columns_to_keep = ['zipcode', 'City']
# Add 'Mean_{year}' columns to the list of columns to keep
start_year = 2015
end_year = 2023
mean_columns = [f'Mean_{year}' for year in range(start_year, end_year + 1)]

# Combine the lists
columns_to_keep.extend(mean_columns)

# Select only the desired columns from ZHVINTA_cleaned
ZORI_selected = ZORI2[columns_to_keep]
```
